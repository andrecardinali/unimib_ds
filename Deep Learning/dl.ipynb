```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
data1 = unpickle("/Users/andreacardinali/unimib_ds/Deep Learning/data/data_batch_1")
data1[b'data'].shape
```

```python
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

fig, axes = plt.subplots(6, 6, figsize = (12,12))
axes = axes.ravel()

for i in range(0, 36):
    index = np.random.randint(0, len(data1[b'data']))
    img = data1[b'data'][index]

    R = img[0:1024].reshape(32, 32)
    G = img[1024:2048].reshape(32, 32)
    B = img[2048:].reshape(32, 32)

    image = np.dstack((R, G, B))

    axes[i].imshow(image)
    label = int(data1[b'labels'][index])
    axes[i].set_title(labels[label])
    axes[i].axis('off')

plt.subplots_adjust(hspace = 0.4)
```
